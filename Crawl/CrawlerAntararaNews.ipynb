```python
import re
import mechanize
import os
import sys
import urllib.request
```

```python
br = mechanize.Browser()
linklist=[]
kamus={}
values = 1
count = 0

day = 1
dayend = 31
mounth = 2
year = 2019
tahun = year
bulan = mounth
tanggal = day
halstart = 1
halend = 15
```

```python
print("Start to execute...")
for tanggal in range(tanggal, dayend):
    print("Crawling data on:", tahun,"/", bulan,"/", tanggal, "|Page", halstart,"-", halend-1)
    i=1
    for i in range(halstart, halend):
        #print("Include page", i)
        if tanggal <10:
            if bulan <10:
                r = br.open("https://www.antaranews.com/indeks/"+ str(tahun)+ "-0" + str(bulan)+ "-0"+ str(tanggal) + "/"+ str(i))
            else:
                r = br.open("https://www.antaranews.com/indeks/"+ str(tahun)+ "-" + str(bulan)+ "-0"+ str(tanggal) + "/"+ str(i))
        else:
            if bulan <10:
                r = br.open("https://www.antaranews.com/indeks/"+ str(tahun)+ "-0" + str(bulan)+ "-"+ str(tanggal) + "/"+ str(i))
            else:
                r = br.open("https://www.antaranews.com/indeks/"+ str(tahun)+ "-" + str(bulan)+ "-"+ str(tanggal) + "/"+ str(i))

        for link in br.links():
            linklist.append(link.url)

        for links in linklist:
            if links not in kamus and re.search(r"/berita/",links):
                kamus[links]=values
                count+=1
                #print(kamus[links])
            else:
                pass
print("Executed successfully")
```

```python
f = open("dataAntara.txt","w")
for x in kamus:
    f.write(x + "\n")
f.close()
```

```python
f = open("dataAntara.txt","r")
count = 1

for link in f:
    print(link)
    urllib.request.urlretrieve(link, "AntaraNews-" + str(count) + ".html")
    count+=1
    print("\nDownloading file: ", count)
print("Finish : ", count)
f.close()
```
